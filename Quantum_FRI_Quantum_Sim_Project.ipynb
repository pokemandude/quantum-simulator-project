```python
# QASM Text compiler

def read_from_file(file_path):
    f = open(file_path, "r")
    # header of QASM file
    f.readline()
    f.readline()
    f.readline()
    # get number of qubits
    line = f.readline()
    num_qubits = re.sub("\D", "", line)
    # get number of classical bits
    line = f.readline()
    num_bits = re.sub("\D", "", line)

    print("number of qubits " + num_qubits)
    print("number of bits " + num_bits)

    f.readline()

    gates_list = []
    # get gates and corresponding qubits
    while True:
        line = f.readline()
        gate_string = ""
        i = 0
        while len(line) and (line[i] != " "):
            gate_string += line[i]
            i += 1
        gate_string += " "
        gate_string += (re.sub("\D", "", line))

        if gate_string[0:7] == "measure":
            gates_list.append("m " + gate_string[8] + " " + gate_string[9])
        else:
            gates_list.append(gate_string)

        if not line:
            break
    gates_list.remove(" ")
    print(gates_list)
    return int(num_qubits), num_bits, gates_list
```

```python
# gate methods
j = complex(0,1) # use j for complex numbers in place of i

def Xgate(qVector):
    newVector = [0,0]
    newVector[0] = 0*qVector[0]+1*qVector[1]
    newVector[1] = 1*qVector[0]+0*qVector[1]
    return newVector

def Ygate(qVector):
    newVector = [0,0]
    newVector[0] = 0*qVector[0]-j*qVector[1]
    newVector[1] = j*qVector[0]+0*qVector[1]
    return newVector

def Zgate(qVector):
    newVector = [0,0]
    newVector[0] = 1*qVector[0]+0*qVector[1]
    newVector[1] = 0*qVector[0]-1*qVector[1]
    return newVector

def Igate(qVector):
    newVector = [0,0]
    newVector[0] = 1*qVector[0]+0*qVector[1]
    newVector[1] = 0*qVector[0]+1*qVector[1]
    return newVector

def Hgate(qVector):
    newVector = [0,0]
    newVector[0] = (1/math.sqrt(2))*(1*qVector[0]+1*qVector[1])
    newVector[1] = (1/math.sqrt(2))*(1*qVector[0]-1*qVector[1])
    return newVector

def Sgate(qVector):
    newVector = [0,0]
    newVector[0] = 1*qVector[0]+0*qVector[1]
    newVector[1] = 0*qVector[0]+j*qVector[1]
    return newVector

def SDGgate(qVector):
    newVector = [0,0]
    newVector[0] = 1*qVector[0]+0*qVector[1]
    newVector[1] = 0*qVector[0]-j*qVector[1]
    return newVector

def Qgate(qVector):
    newVector = [0,0]
    newVector[0] = (1/math.sqrt(2))*(1*qVector[0]+1*qVector[1])
    newVector[1] = (1/math.sqrt(2))*(j*qVector[0]-j*qVector[1])
    return newVector

def QDGgate(qVector):
    newVector = [0,0]
    newVector[0] = (1/math.sqrt(2))*(1*qVector[0]-j*qVector[1])
    newVector[1] = (1/math.sqrt(2))*(1*qVector[0]+j*qVector[1])
    return newVector

def Tgate(qVector): # do this with complex number
    newVector = [0,0]
    newVector[0] = 1*qVector[0]+0*qVector[1]
    newVector[1] = 0*qVector[0]+(e**(j*math.pi/4))*qVector[1]
    return newVector

def TDGgate(qVector): # do this with complex number
    newVector = [0,0]
    newVector[0] = 1*qVector[0]+0*qVector[1]
    newVector[1] = 0*qVector[0]+(e**(-j*math.pi/4))*qVector[1]
    return newVector

#add more gates T-T
```
